```python
import pandas as pd
from pandas import read_csv
# import str
import numpy as np
```

```python
df = pd.read_csv('data/hanzidb_hsk_characters_raw.csv')
df.head()
```

```python
# Cleaning the radical column
split_data = df['radical'].str.split('\xa0', expand=True)
df['radical_char'] = split_data[0]
df['radical_index'] = split_data[1]
df.drop(columns=['radical'], inplace=True)
```

```python
df['radical_index'] = pd.to_numeric(df['radical_index'], errors='coerce')
```

```python
df = df[['char', 'pinyin', 'definition', 'radical_char', 'radical_index', 'stroke_count', 'hsk_level', 'general_standard', 'frequency_rank']]
# Clean stroke count column
df['stroke_count'] = df['stroke_count'].astype(str).str.split().str[0]
df['stroke_count'] = pd.to_numeric(df['stroke_count'], errors='coerce')
```

```python
# Export cleaned data to new CSV
df.to_csv('data/hanzidb_hsk_characters_cleaned.csv', index=False)
```
